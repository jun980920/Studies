```python
import folium
```

```python
long=df_gy['경도'].mean()
lat=df_gy['위도'].mean()
```

```python
m=folium.Map(location=(lat, long), zoom_start=12)
for i in df_gy.index:
    long=df_gy.loc[i,'경도']
    lat=df_gy.loc[i,'위도']
    tooltip=df_gy.loc[i,'공원명']+df_gy.loc[i, '소재지도로명주소']
    if i in list(k.index):
        color='blue'
    else:
        color='green'
    folium.Marker([lat, long], tooltip=tooltip, icon=folium.Icon(color=color)).add_to(m)
m
```
