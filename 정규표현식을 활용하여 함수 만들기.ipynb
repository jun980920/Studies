```python
def get_string(text):
    if pd.isna(text):
        return text
    text=re.sub("[\d\.\-(nan)\[\]\/']", '', text)
    text=re.sub(' ','',text)
    text=re.sub(',{2,}',',',text)
    return text
text="[1. '화장실' 5, 2. 주차장, -, nan]"
get_string(text)
```

```python
get_string(np.nan)
```

```python
df.head()
```

```python
df['편익시설']=df['공원보유시설(편익시설)'].map(get_string)
pd.set_option("max_rows", None)
df['편익시설']
```
